```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("daily_temperature.csv")

# Preview dataset
print(data.head())

# Plot Line Chart
plt.figure(figsize=(10,6))
plt.plot(data['Day'], data['Temperature'], marker='o', linestyle='-', color='b')

# Add title, labels, and grid
plt.title("Daily Temperature Changes Over a Month")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.grid(True, linestyle="--", alpha=0.7)

# Highlight hottest and coldest days
hottest_day = data.loc[data['Temperature'].idxmax()]
coldest_day = data.loc[data['Temperature'].idxmin()]

plt.scatter(hottest_day['Day'], hottest_day['Temperature'], color='red', s=100, label=f"Hottest Day ({hottest_day['Day']})")
plt.scatter(coldest_day['Day'], coldest_day['Temperature'], color='blue', s=100, label=f"Coldest Day ({coldest_day['Day']})")

plt.legend()
plt.tight_layout()
plt.show()

print(f"🔥 Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}°C")
print(f"❄️ Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("sales_data.csv")

# Plot Line Chart
plt.figure(figsize=(8,5))
plt.plot(data['Quarter'], data['Sales'], marker='o', linestyle='-', color='green')

# Add title, labels, and grid
plt.title("Quarterly Sales Performance")
plt.xlabel("Quarter")
plt.ylabel("Sales ($)")
plt.grid(True, linestyle="--", alpha=0.7)

# Highlight peak sales
peak = data.loc[data['Sales'].idxmax()]
plt.scatter(peak['Quarter'], peak['Sales'], color='red', s=120, label=f"Peak Sales ({peak['Quarter']})")

plt.legend()
plt.tight_layout()
plt.show()

print(f"🚀 Peak Sales: {peak['Quarter']} with ${peak['Sales']:,}")
```
