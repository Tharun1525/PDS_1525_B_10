```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 1. Load dataset
# Expected CSV format: Country,Population
df = pd.read_csv("population.csv")

# 2. Sort and select top 5 countries
top5 = df.sort_values(by="Population", ascending=False).head(5)

# 3. Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(top5["Country"], top5["Population"], color="skyblue", edgecolor="black")

# Format y-axis (millions/billions)
def human_format(x, pos):
    if x >= 1_000_000_000:
        return f'{x/1_000_000_000:.1f}B'
    elif x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    else:
        return str(int(x))

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(human_format))

# 4. Labels & title
plt.title("Top 5 Most Populous Countries", fontsize=16, weight="bold")
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)

# Show values on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             human_format(height, None),
             ha='center', va='bottom', fontsize=10, weight="bold")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv("movies_genre.csv")

# 2. Plot bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(df["Genre"], df["MovieCount"], color="teal", edgecolor="black")

# 3. Formatting
plt.title("Number of Movies Produced by Genre", fontsize=16, weight="bold")
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.xticks(rotation=45, ha="right")

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 30,   # +30 for spacing
             str(height),
             ha='center', va='bottom', fontsize=9, weight="bold")

plt.tight_layout()
plt.show()
```
