```python
import pandas as pd

# --- Create Products dataset ---
products = pd.DataFrame({
    "product_id": [101, 102, 103, 104, 105],
    "product_name": ["Laptop", "Smartphone", "Desk Chair", "Table", "Headphones"],
    "category": ["Electronics", "Electronics", "Furniture", "Furniture", "Electronics"],
    "price": [800, 600, 120, 200, 150]
})

# --- Create Sales dataset ---
sales = pd.DataFrame({
    "sale_id": [1, 2, 3, 4, 5, 6],
    "product_id": [101, 102, 103, 104, 105, 106],
    "quantity": [10, 15, 20, 5, 25, 12]
})

print("Products Dataset:\n", products, "\n")
print("Sales Dataset:\n", sales, "\n")

# --- Inner Join ---
inner_merge = pd.merge(sales, products, on="product_id", how="inner")
print("Inner Join Result:\n", inner_merge, "\n")

# --- Outer Join ---
outer_merge = pd.merge(sales, products, on="product_id", how="outer")
print("Outer Join Result:\n", outer_merge, "\n")

# --- Compute Statistics ---
# Total revenue per product
inner_merge["revenue"] = inner_merge["quantity"] * inner_merge["price"]

total_revenue = inner_merge["revenue"].sum()
avg_price = products["price"].mean()
product_sales = inner_merge.groupby("product_name")["revenue"].sum()
category_sales = inner_merge.groupby("category")["revenue"].sum()

best_category = category_sales.idxmax()

print("üìä Total Revenue: $", total_revenue)
print("üìä Average Product Price: $", round(avg_price, 2))
print("üìä Product-wise Sales:\n", product_sales, "\n")
print("üìä Category-wise Sales:\n", category_sales, "\n")
print("üèÜ Best-selling Category:", best_category)
```

```python
import pandas as pd

# --- Teachers dataset ---
teachers = pd.DataFrame({
    "teacher_id": [1, 2, 3, 4, 5],
    "teacher_name": ["Dr. Smith", "Dr. Johnson", "Dr. Lee", "Dr. Brown", "Dr. Adams"],
    "department": ["Computer Science", "Mathematics", "Physics", "Computer Science", "Mathematics"]
})

# --- Courses dataset ---
courses = pd.DataFrame({
    "course_id": [101,102,103,104,105,106,107,108,109,110,111],
    "course_name": ["Algorithms", "Data Structures", "Calculus", "Linear Algebra",
                    "Quantum Mechanics", "Databases", "Operating Systems",
                    "Statistics", "Machine Learning", "Astrophysics", "Topology"],
    "teacher_id": [1,1,2,2,3,4,4,5,1,3,6]
})

print("Teachers Dataset:\n", teachers, "\n")
print("Courses Dataset:\n", courses, "\n")

# --- Inner Join (only matching teacher_id) ---
inner_merge = pd.merge(courses, teachers, on="teacher_id", how="inner")
print("Inner Join Result:\n", inner_merge, "\n")

# --- Left Join (all courses, keep unmatched teachers as NaN) ---
left_merge = pd.merge(courses, teachers, on="teacher_id", how="left")
print("Left Join Result:\n", left_merge, "\n")

# --- Compute Statistics ---
# Average courses per teacher
courses_per_teacher = inner_merge.groupby("teacher_name")["course_id"].count()
avg_courses = courses_per_teacher.mean()

# Workload per department
workload_by_dept = inner_merge.groupby("department")["course_id"].count()

print("üìä Courses per Teacher:\n", courses_per_teacher, "\n")
print("üìä Average Courses per Teacher:", round(avg_courses, 2))
print("üìä Workload by Department:\n", workload_by_dept, "\n")
```
